```python
import numpy as np
import matplotlib.pyplot as plt
# plt.style.use(["seaborn-v0_8", "paper.mplstyle"])
# plt.rcParams["lines.linewidth"] = 2
```

La ecuaci√≥n diferencial es
$$
dy/dx = y(1-y)
$$
o
$$
1\cdot dy/dx = y(1-y)
$$

de donde se obtiene
$$
u = 1 \\
v = y(1-y)
$$

```python
# y' = y(1-y)
x = np.linspace(-1.5,1.5,11)
y = x.copy()
x,y = np.meshgrid(x,y)
u = x*0 + 1
v = y*(1-y)
plt.figure(figsize=(4,4))
plt.streamplot(x,y,u,v)
plt.axis("equal")
plt.quiver(x,y,u,v, color="red")
plt.axis("equal")
```

Si se considera la ecuaci√≥n de circulos concentricos
$$
x^2 + y^2 = c^2
$$

y se deriva respecto a $x$, entonces se obtiene
$$
2x + 2y\frac{dy}{dx} = 0 
$$

despejando se obtiene la ecuaci√≥n diferencial
$$
\frac{dy}{dx} = -\frac{x}{y}
$$

para evitar diviciones por cero, es mejor despejar para no tener divisiones en $u$ o en $v$
$$
y\frac{dy}{dx} = -x
$$

de donde se obtiene
$$
u = y \\
v = -x
$$

Otra forma de verlo es que la pendiente $dy/dx$ es el vector que define la direcci√≥n de las curvas, en direcci√≥n $v/u$

```python
x = np.linspace(-5,5,11)
y = x.copy()
x,y = np.meshgrid(x,y)
u = y
v = -x
plt.figure(figsize=(4,4))
plt.streamplot(x,y,u,v)
plt.axis("equal")
plt.quiver(x,y,u,v, color="red")
plt.axis("equal")
```
